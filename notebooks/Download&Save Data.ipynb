# Load Yahoo Financials Library

```python
from yahoofinancials import YahooFinancials
import pandas as pd
```

# Load TOP40 Companies

```python
tickers = pd.read_csv('../data/external/jse_top_40.csv')
tickers['Symbol'] +='.JO'
tickers.columns = ['Company_Name','URL_Description','Symbol','Sector']
```

# Annual Cash Statement History

```python
companies = pd.DataFrame()
for i in range(len(tickers)):
        symbol = tickers.iloc[i]['Symbol']
        
        symbol = symbol
        
        try:
            df = pd.DataFrame()
            data = []
            [data.append(pd.DataFrame.from_dict(i).T) for i in YahooFinancials(symbol).get_financial_stmts('annual', 'cash')['cashflowStatementHistory'][symbol]]
    
            for i in data:
                df = pd.concat([df,i])
                
            df['Ticker'] = symbol
            
            companies = pd.concat([companies, df])
        except:
            pass

companies['Date'] = pd.to_datetime(companies.index)
companies        
        
companies.head()
```

# Annual Balance Sheet Statement History

```python
companies = pd.DataFrame()
for i in range(len(tickers)):
        symbol = tickers.iloc[i]['Symbol']
        
        symbol = symbol
        
        try:
            df = pd.DataFrame()
            data = []
            [data.append(pd.DataFrame.from_dict(i).T) for i in YahooFinancials(symbol).get_financial_stmts('annual', 'balance')['balanceSheetHistory'][symbol]]
    
            for i in data:
                df = pd.concat([df,i])
                
            df['Ticker'] = symbol
            
            companies = pd.concat([companies, df])
        except:
            pass

companies['Date'] = pd.to_datetime(companies.index)

balance = companies.copy()

balance = balance.reset_index()    

balance.to_csv('../data/raw/TOP40_Balance_Annual.csv', index=False)

balance.head()
```

# Annual Income Statement History

```python
companies = pd.DataFrame()
for i in range(len(tickers)):
        symbol = tickers.iloc[i]['ari-tbl-col-1']
        
        symbol = symbol+'.JO'
        
        try:
            df = pd.DataFrame()
            data = []
            [data.append(pd.DataFrame.from_dict(i).T) for i in YahooFinancials(symbol).get_financial_stmts('annual', 'income')['incomeStatementHistory'][symbol]]
    
            for i in data:
                df = pd.concat([df,i])
                
            df['Ticker'] = symbol
            
            companies = pd.concat([companies, df])
        except:
            pass

companies['Date'] = pd.to_datetime(companies.index)

income = companies.copy()
        
income = income.reset_index()    

income.to_csv('../data/raw/TOP40_Income_Annual.csv', index=False)
    
income.head()
```

# Download Historical Stock Price

```python
daily_bank_stock_prices = []
#[daily_bank_stock_prices.append(YahooFinancials(i).get_historical_price_data('2008-01-01', '2019-04-20', 'weekly')) for i in tickers.Symbol.unique()]

for i in tickers.Symbol.unique():
    
    try:
        daily_bank_stock_prices.append(YahooFinancials(i).get_historical_price_data('2008-01-01', '2019-04-20', 'weekly'))

    except:
        print('Could not find:',i)
        pass

```

# Format JSON to CSV

```python
prices=pd.DataFrame()

for i in daily_bank_stock_prices:
    for j in i.keys():
            try:
                #print(j)
                temp = pd.DataFrame.from_dict(i[j]['prices'])
                temp['Symbol']=j
                prices=pd.concat([prices,temp])
            except:
                pass

prices.head()

prices.to_csv('../data/raw/TOP40_Historical_Sales.csv',index=False)
```

```python
prices['date'] = pd.to_datetime(prices['date'])

prices['date'] = prices['formatted_date']
```
